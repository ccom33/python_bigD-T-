```python
from tqdm import tqdm_notebook

present_candi_text = []

for n in tqdm_notebook(range(1, 10000, 10)):
    tmp1 ='https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='
    
    quote = rep.quote_plus('여친 선물')
    
    url = tmp1 + quote
    
    req = urllib.request.Request(url, headers={'User-Agent':'Mozilla/5.0'})
    
    response = urllib.request.urlopen(req)
    
    soup = BeautifulSoup(response, 'html.parser')
    
    tmp = soup.find_all('dl')
    
    for line in tmp:
        present_candi_text.append(line.text)
        
    time.sleep (0.1)
```

```python
present_candi_text
```

```python
from konlpy.tag import Okt
o = Okt()
import nltk
```

```python
present_text = ''

for each_line in present_candi_text[:10000]:
    present_text = present_text + each_line + '\n'
```

```python
tokens_ko = o.morphs(present_text)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='여자친구 선물')
print(len(ko.tokens))
print(len(set(ko.tokens))) 
```

```python
ko = nltk.Text(tokens_ko, name='여자친구 선물')
ko.vocab().most_common(100)
```
